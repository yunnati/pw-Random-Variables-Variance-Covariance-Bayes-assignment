Q. 1: NoSQL vs Relational Databases a) What are the key differences between relational databases and MongoDB? b) When is MongoDB preferred over SQL systems? c) Explain schema flexibility and its risks. d) How does embedding vs referencing impact performance?

- a) What are the key differences between relational databases and MongoDB?

     - Relational databases store data in tables consisting of rows and columns, whereas MongoDB stores data in collections as JSON-like documents (BSON).
     - SQL databases use a fixed schema, meaning every record follows the same structure, while MongoDB has a flexible schema that allows documents in the same collection to have different fields.
     - Relational databases use SQL for querying and are well-suited for structured data with complex relationships. MongoDB uses MongoDB Query Language (MQL) and is better suited for semi-structured or unstructured data.
     - SQL databases commonly use JOIN operations, while MongoDB generally uses embedding or referencing to represent relationships.

- b) When is MongoDB preferred over SQL systems?

     - MongoDB is preferred when the application requires a flexible schema, handles large volumes of semi-structured or unstructured data, or needs to scale horizontally.
     - It is widely used in applications such as e-commerce platforms, social media applications, etc. where the data structure changes frequently.

- c) Explain schema flexibility and its risks.

     - Schema flexibility means documents in the same MongoDB collection are not required to have identical fields. This allows developers to easily add or modify fields without changing the entire database structure. However, it can also lead to inconsistent data, missing fields, duplicate information, and the need for additional validation in the application.

- d) How does embedding vs referencing impact performance?

     - Embedding stores related data within a single document, which improves read performance because all related information is retrieved in one query. However, it may increase document size and duplicate data.
     - Referencing stores related data in separate documents connected through IDs. This reduces duplication and is suitable for large or many-to-many relationships, but retrieving related data may require additional queries, making reads slightly slower.


Q. 2: Indexing, Sharding & Replication a) How do indexes improve MongoDB query performance? b) What is sharding? When is it required? c) What is replication and how does it ensure high availability? d) How does improper shard key selection impact performance?

- a) How do indexes improve MongoDB query performance?

    - Indexes improve query performance by allowing MongoDB to quickly locate documents without scanning the entire collection. They store indexed fields in a sorted structure, making searching, sorting, and filtering much faster.

- b) What is sharding? When is it required?

    - Sharding is the process of distributing data across multiple servers or shards. It is required when the database becomes too large for a single server or when the application receives very high read and write traffic. Sharding improves scalability and balances the workload across multiple machines.

- c) What is replication and how does it ensure high availability?

    - Replication creates multiple copies of the same data on different servers. One server acts as the primary node, while the others act as secondary nodes.
    - If the primary server fails, one of the secondary servers automatically becomes the new primary. This ensures continuous availability of the database and protects against data loss.

- d) How does improper shard key selection impact performance?

    - An improper shard key can distribute data unevenly among shards, causing some servers to become overloaded while others remain underutilized. This results in slower queries, poor load balancing, hotspots, and reduced overall performance.

Q. 3: Descriptive Statistics & Distributions a) Difference between mean, median, and variance b) What does high variance indicate in customer spending? c) Difference between normal distribution and skewed distribution d) Why do business metrics often show right-skewed distributions?

- a) Difference between mean, median, and variance.

    - The mean is the arithmetic average of all values.
    - The median is the middle value after arranging the data in ascending order and is less affected by extreme values.
    - Variance measures how far the data values are spread from the mean. A higher variance indicates greater variability in the data.

- b) What does high variance indicate in customer spending?

    - High variance indicates that customer spending differs significantly. Some customers spend very little, while others spend much more. This suggests inconsistent purchasing behavior among customers.

- c) Difference between normal distribution and skewed distribution.

    - A normal distribution is symmetrical with most values concentrated around the mean, median, and mode are approximately equal.
    - A skewed distribution is asymmetrical, where one side has a longer tail than the other. In a right-skewed distribution, a few very large values pull the mean to the right.

- d) Why do business metrics often show right-skewed distributions?

    - Business metrics such as customer spending, sales, and income are often right-skewed because most customers spend moderate amounts, while a small number of customers spend exceptionally large amounts. These high values create a long right tail.

- a) What is a random variable?

    - A random variable is a numerical value assigned to the outcome of a random experiment. For example, the number of orders received in a day is a random variable because it changes from day to day.

- b) Difference between discrete and continuous random variables.

    - A discrete random variable takes countable values such as the number of customers or products sold. A continuous random variable takes any value within a range, such as delivery time, customer waiting time, or product weight.

- c) Explain covariance and its interpretation.

    - Covariance measures how two variables change together. A positive covariance indicates that both variables tend to increase or decrease together.
    - A negative covariance means that when one variable increases, the other tends to decrease. A covariance close to zero indicates little or no linear relationship.

- d) Explain Bayes' theorem in a business decision-making context.

    - Bayes' theorem is used to calculate the probability of an event based on new information. In business, it helps estimate probabilities such as customer churn, fraud detection, or product purchase likelihood by combining prior knowledge with newly observed data.

Q. 8: Suppose: ● 5% of customers churn ● 70% of churned customers had declining order frequency ● 15% of non-churned customers had declining order frequency If a customer shows declining frequency, compute probability that they will churn.

- Given:

    - P(Churn) = 0.05
    - P(Declining | Churn) = 0.70
    - P(Declining | Not Churn) = 0.15

- First calculate the probability of declining order frequency:

      P(Declining) = (0.70 × 0.05) + (0.15 × 0.95)

      = 0.035 + 0.1425

      = 0.1775

      Using Bayes' Theorem:

      P(Churn | Declining) = (0.70 × 0.05) / 0.1775

      = 0.035 / 0.1775

      ≈ 0.1972

    - Therefore, the probability that a customer will churn given declining order frequency is approximately 19.7%.

Q. 9: Explain: 1. Why CLT allows us to use normal approximation 2. Why it is important in A/B testing 3. What happens if sample size is very small?

- 1. Why does CLT allow us to use normal approximation?

    - The Central Limit Theorem states that when the sample size is sufficiently large, the distribution of the sample mean becomes approximately normal, regardless of the original population distribution. This allows normal distribution methods to be applied in statistical analysis.

- 2. Why is it important in A/B testing?

    - CLT enables researchers to compare the average performance of two groups using confidence intervals and hypothesis tests. It forms the statistical foundation of most A/B testing methods used in business and experimentation.

- 3. What happens if the sample size is very small?

    - When the sample size is very small, the sampling distribution may not be normally distributed. Statistical estimates become less reliable, confidence intervals become wider, and hypothesis tests have lower accuracy.

Q. 10:The company claims: “Average monthly spending is ₹5000.” 1. How would you statistically test this claim? 2. Define null and alternative hypothesis 3. What type of error is more dangerous in business — Type I or Type II?

- 1. How would you statistically test this claim?

    - To test the claim that the average monthly spending is ₹5000, collect a random sample of customer spending and perform a one-sample t-test (or a z-test if the population standard deviation is known). Calculate the test statistic and compare the p-value with the chosen significance level (usually 0.05). Based on the result, either reject or fail to reject the claim.

- 2. Define null and alternative hypothesis.

      The null hypothesis (H₀) states that the average monthly spending is equal to ₹5000.

       H₀: μ = ₹5000

      The alternative hypothesis (H₁) states that the average monthly spending is different from ₹5000.

      H₁: μ ≠ ₹5000

- 3. What type of error is more dangerous in business—Type I or Type II?

    - Type II error is often more dangerous because it means failing to detect a real problem or opportunity. For example, failing to identify customers who are likely to churn or missing fraudulent transactions can result in significant financial losses.
